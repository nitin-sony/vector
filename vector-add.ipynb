```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors u and v
u = np.array([4, 1])
v = np.array([1, 3])

# Compute sum vector (diagonal of the parallelogram)
sum_vector = u + v

# Create the plot
fig, ax = plt.subplots(figsize=(6,6))

# Plot vectors u and v from origin (0,0)
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\mathbf{u}$')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\mathbf{v}$')

# Plot the sum vector (diagonal of the parallelogram)
ax.quiver(0, 0, sum_vector[0], sum_vector[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\mathbf{u} + \mathbf{v}$')

# Draw the parallelogram by adding translated vectors
ax.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5, linestyle='dashed')
ax.quiver(v[0], v[1], u[0], u[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5, linestyle='dashed')

# Set axis limits
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 6)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True, linestyle='--', alpha=0.6)

# Labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Vector Addition and Parallelogram Law")
ax.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure
fig, ax = plt.subplots(figsize=(6,6))

# Define vectors u and v
a = np.array([5, 7])
b = np.array([2, 4])

# Plot the points
ax.scatter(*a, color='b', label="Vector A (5,7)", s=100)
ax.scatter(*b, color='g', label="Vector B (2,4)", s=100)

# Draw the L2 (Euclidean) distance as a direct line
ax.plot([a[0], b[0]], [a[1], b[1]], 'r--', label="L2 Distance (Euclidean)")

# Draw the L1 (Manhattan) distance as a right-angle path
ax.plot([a[0], a[0]], [a[1], b[1]], 'gray', linestyle='dotted')
ax.plot([a[0], b[0]], [b[1], b[1]], 'gray', linestyle='dotted', label="L1 Distance (Manhattan)")

# Set axis limits
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)

# Labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("L1 vs L2 Distance Visualization")

# Grid and legend
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend()

# Show plot
plt.show()
```
